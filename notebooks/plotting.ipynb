```python
colors = ['#ff0000', '#ffa000', '#00aaff']
markers = ['D', 'p', '*']
labels= ['Adam', 'PGD','Inflate', ]
```

```python
# means = [nd-array, nd-array, nd-array]
def plotMethod(means, stds, ax, title):
    for i, (mean, std) in enumerate(zip(means, stds)):
        ax.errorbar(mean, np.arange(12,0,-4)-i, xerr=std, fmt='none', color='black', capsize=4, elinewidth=1.5, capthick=1.5)
        ax.scatter(mean, np.arange(12,0,-4)-i, marker=markers[i], s=16, color=colors[i], zorder=10, label=labels[i])

    ax.legend(loc='best')
    #ax.set_title(title)
    plt.rcParams.update({'ytick.right': False})
    plt.rcParams.update({'ytick.left': False})
    ax.set_yticks((2,3,4,6,7,8,10,11,12), labels=['','CAA','', '','FGSM','', '','Clean', ''])


```

```python
fig, ax = plt.subplots()
plotMethod(means_fashionMNIST, stds_fashionMNIST, ax, 'Mean accuracies on fashionMNIST')


```
